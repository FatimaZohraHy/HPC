```python
import matplotlib.pyplot as plt
import numpy as np

# Execution times with and without MPI_Barrier for master and slave nodes
execution_times_with_barrier = [0.000560, 0.000083]  # [Master, Slave]
execution_times_without_barrier = [0.000001, 0.000001]  # [Master, Slave]

# Labels for the nodes
nodes = ['Master Node', 'Slave Node']

# X-axis positions for the bars
x = np.arange(len(nodes))

# Bar width
width = 0.35

# Create the bar plot
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, execution_times_with_barrier, width, label='With Barrier', color='b')
bars2 = ax.bar(x + width/2, execution_times_without_barrier, width, label='Without Barrier', color='orange')

# Labeling
ax.set_xlabel("Node Type")
ax.set_ylabel("Execution Time (seconds)")
ax.set_title("Execution Time Comparison with and without MPI_Barrier")
ax.set_xticks(x)
ax.set_xticklabels(nodes)
ax.legend()

# Display values on top of bars
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.6f}',  # Formatting the time for better readability
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # Offset text a bit
                textcoords="offset points",
                ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
#execution time of multiplication two matrices in sequential & parallel version
```

```python
import matplotlib.pyplot as plt

# Execution times for sequential and parallel versions
execution_times = {
    'Sequential': 0.000002,
    'Parallel (2 processes)': 0.000254
}

# Separate the keys and values for plotting
versions = list(execution_times.keys())
times = list(execution_times.values())

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(versions, times, color=['blue', 'orange'])
plt.xlabel("Version")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time Comparison: Sequential vs Parallel Matrix Multiplication")
plt.yscale('log')  # Use a log scale to better visualize the difference

# Display values on top of bars
for i, v in enumerate(times):
    plt.text(i, v + 0.0000001, f"{v:.7f}", ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Execution times for sequential and parallel versions
execution_times = {
    'Sequential': 0.300929,
    'Parallel (4 processes)': 0.084453
}

# Separate the keys and values for plotting
versions = list(execution_times.keys())
times = list(execution_times.values())

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(versions, times, color=['blue', 'orange'])
plt.xlabel("Version")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time Comparison: Sequential vs Parallel Integral Computation")
plt.yscale('linear')  # Use a linear scale for this comparison

# Display values on top of bars
for i, v in enumerate(times):
    plt.text(i, v + 0.002, f"{v:.6f}", ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Number of processors and their corresponding execution times
processors = list(range(2, 11))  # Number of processors from 2 to 10
execution_times = [
    0.158332,  # for np=2
    0.106550,  # for np=3
    0.084453,  # for np=4
    0.059686,  # for np=5
    0.060734,  # for np=6
    0.064798,  # for np=7
    0.048955,  # for np=8
    0.034407,  # for np=9
    0.035775   # for np=10
]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(processors, execution_times, marker='o', linestyle='-', color='b')
plt.title('Execution Time of Parallel Integral Computation vs. Number of Processors')
plt.xlabel('Number of Processors')
plt.ylabel('Execution Time (seconds)')
plt.xticks(processors)  # Set x-ticks to match the number of processors
plt.grid(True)
plt.ylim(0, max(execution_times) * 1.1)  # Set y-limit slightly above max time

# Annotate the points with their execution times
for i, time in enumerate(execution_times):
    plt.text(processors[i], time + 0.002, f"{time:.6f}", ha='center')

plt.show()
```
